```python
import pandas as pd
import numpy as np
```

```python
data_df = pd.read_csv('Assets/SPY.csv')

data_df
```

```python
data_df['Year'] = data_df['Date'].apply(lambda x: int(x.split('-')[0]))

data_df['Month'] = data_df['Date'].apply(lambda x: int(x.split('-')[1]))

years = data_df['Year'].unique().tolist()

months = data_df['Month'].unique().tolist()

return_s = []

dates = []

for year in years:
    
    for month in months:

        first_close = data_df.loc[(data_df['Year'] == year) & (data_df['Month'] == month)]['Close'].tolist()[0]

        last_close = data_df.loc[(data_df['Year'] == year) & (data_df['Month'] == month)]['Close'].tolist()[-1]

        return_y = (last_close - first_close) / first_close
    
    
    
        return_s.append(return_y)
        
        dates.append(f'{year}/{month}')
    
testing_df = pd.DataFrame([dates, return_s]).transpose()

testing_df.columns=['Date', 'SPY']
```

```python
bonds = ["^TNX", "^TYX", "^FVX", "^IRX"]
```

```python
pd.read_csv(f'Assets/{bond}.csv')
```

```python
for bond in bonds:
    
    tnx = pd.read_csv(f'Assets/{bond}.csv')

    tnx['Year'] = tnx['Date'].apply(lambda x: int(x.split('-')[0]))

    tnx['Month'] = tnx['Date'].apply(lambda x: int(x.split('-')[1]))

    years = tnx['Year'].unique().tolist()

    months = tnx['Month'].unique().tolist()

    return_s = []

    dates = []

    for year in years:

        for month in months:
            
            try:

                first_close = tnx.loc[(tnx['Year'] == year) & (tnx['Month'] == month)]['Close'].tolist()[0]
                
            except:

                dates.append(np.nan)
                
                return_s.append(np.nan)
                
                continue

            last_close = tnx.loc[(tnx['Year'] == year) & (tnx['Month'] == month)]['Close'].tolist()[-1]

            avg = (first_close + last_close) / 200    

            return_y = ((1 + avg) ** (1/12)) - 1



            return_s.append(return_y)

            dates.append(f'{year}/{month}')

    temp = pd.DataFrame([dates, return_s]).transpose()

    temp.columns=['Date', bond]

    testing_df = testing_df.merge(temp, on='Date', how='inner')
    
```
